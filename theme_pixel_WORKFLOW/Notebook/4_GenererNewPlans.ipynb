```python
##Simple test of similraty to calculate score of old study plan ( Not admitted)
from transformers import BertTokenizer, BertModel
import torch.nn.functional as F

# Load pre-trained tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Define two example lists
list1 = listaa
list2 = lista2

# Tokenize each item in the lists
tokens1 = [tokenizer(item, padding=True, truncation=True, return_tensors='pt') for item in list1]
tokens2 = [tokenizer(item, padding=True, truncation=True, return_tensors='pt') for item in list2]

# Get the embeddings for each item in the lists
embeddings1 = [model(**token).last_hidden_state.mean(dim=1) for token in tokens1]
embeddings2 = [model(**token).last_hidden_state.mean(dim=1) for token in tokens2]

# Calculate the semantic similarity between each pair of embeddings
similarity_scores = []
for emb1 in embeddings1:
    for emb2 in embeddings2:
        sim = F.cosine_similarity(emb1, emb2, dim=1)
        similarity_scores.append(sim.item())

# Aggregate the similarity scores
overall_similarity = sum(similarity_scores) / len(similarity_scores)

print("Overall similarity score:", overall_similarity)
```
